```python
import pandas as pd
path="C:\\Users\\Md Murshid Alam\\Desktop\\ml\\turnover.csv"
df=pd.read_csv(path,engine='python',encoding='latin1')
print(df)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
print('Categorical columns: ')
for col in df.columns:
    if df[col].dtype == 'object':
        values = df[col].value_counts()
        values = dict(values)
        
        print(str(col))
        label = LabelEncoder()
        label = label.fit(df[col])
        df[col] = label.transform(df[col].astype(str))
        
        new_values = df[col].value_counts()
        new_values = dict(new_values)
        
        value_dict = {}
        i=0
        for key in values:
            value_dict[key] = list(new_values)[i]
            i+= 1
        print(value_dict)
```
